```python
import matplotlib.pyplot as plt
from strucpara.analysis import GrooveAgent
rootfolder = '/home/ytcdata/x3dna_data'
```

### Part 1: Initialize

```python
time_interval = '0_1us' # '0_1us', '1_2us', '2_3us', '3_4us', '4_5us'
bp_agent = GrooveAgent(rootfolder, time_interval)
```

### Part 2: Histogram for parameters

```python
figsize = (12, 4)
bins = 50
sele_para = 'major_gw_pp' #'major_gw_pp', 'major_gw_refined', 'minor_gw_pp', 'minor_gw_refined'
xlines = range(12,28,2)
xlim = (12,28)
ylim = None

fig, axes = bp_agent.histogram_three_groups(figsize, sele_para, bins, xlines, xlim, ylim)
plt.tight_layout()
#plt.savefig(f'{sele_para}.svg')
plt.show()
```
