### Batch Run

```python
hosts = ['tgtg_21mer', 'ctct_21mer']
time_intervals = ['0_1us', '1_2us', '2_3us', '3_4us', '4_5us']
type_na = 'bdna+bdna'
n_bp = 21
for host in hosts:
    for time_interval in time_intervals:
        bp_agent = ZpAgent(rootfolder, host, time_interval, type_na, n_bp)
        bp_agent.extract_parameters()
        bp_agent.convert_dat_to_csv()
```
