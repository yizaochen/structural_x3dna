```python
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(ncols=3, nrows=1, figsize=(8,4))

ax = axes[0]
ax.set_title("0")

ax = axes[1]
ax.set_title("1")

ax = axes[2]
ax.set_title("2")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(ncols=2, nrows=2, figsize=(8,8))

ax = axes[0,0]
ax.set_title("0,0")

ax = axes[0,1]
ax.set_title("0,1")

ax = axes[1,0]
ax.set_title("1,0")

ax = axes[1,1]
ax.set_title("1,1")

plt.tight_layout()
plt.show()
```
