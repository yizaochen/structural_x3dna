```python
import matplotlib.pyplot as plt
import numpy as np
from strucpara.analysis import PhaseChiDeltaAgent
rootfolder = '/home/ytcdata/x3dna_data'
```

### Part 1: Initialize

```python
p_agent = PhaseChiDeltaAgent(rootfolder)
```

### Part 2: Four systems, 0-5$\mu$s, Phase Angle

```python
figsize = (12,12)
outer_wspace = 0.22
outer_hspace = 0.22
inner_wspace = 0
inner_hspace = 0.15
bins = 100
xlims = None
ylims = (0, 0.025)
xlines = range(0,360,50)
ylines = None

d_axes = p_agent.histogram_phase_angle_four_systems(figsize, outer_wspace, outer_hspace, inner_wspace, inner_hspace, bins, xlims, ylims, xlines, ylines)

#plt.tight_layout()
#plt.savefig("/home/yizaochen/Desktop/drawzone_temp/phase_angle.png", dpi=150)
plt.show()
```

### Part 3: A-junction Central Region Phase Angle

```python
host = 'tat_21mer'
parameter = 'phase2'
data_junction = p_agent.get_data_central_A_junction(parameter, host)
data_nonjunction = p_agent.get_data_central_A_junction_non_junction_part(parameter, host)
```

```python
bins = 100
xlines = range(0,360,50)
ylines = np.arange(0.005, 0.026, 0.005)
fig, axes = plt.subplots(nrows=2, figsize=(8,8), sharey=True, facecolor="white")

ax = axes[0]
ax.hist(data_junction, bins=bins, density=True, color='green', alpha=0.4)
for xvalue in xlines:
    ax.axvline(xvalue, color="grey", alpha=0.1)
for yvalue in ylines:
    ax.axhline(yvalue, color="grey", alpha=0.1)
ax.set_xlabel("Phase Angle")
ax.set_ylabel("P")
ax.set_title('A-junction Strand2 Junction Part')

ax = axes[1]
ax.hist(data_nonjunction, bins=bins, density=True, color='green', alpha=0.4)
for xvalue in xlines:
    ax.axvline(xvalue, color="grey", alpha=0.1)
for yvalue in ylines:
    ax.axhline(yvalue, color="grey", alpha=0.1)
ax.set_xlabel("Phase Angle")
ax.set_ylabel("P")
ax.set_title('A-junction Strand2 Non-Junction Part')

plt.tight_layout()
#plt.savefig("/home/yizaochen/Desktop/drawzone_temp/A_junction_phase_angle.png", dpi=150, transparent=False)
plt.show()
```

### Part 4: $\chi$-$\delta$

```python
figsize = (12,12)
outer_wspace = 0.22
outer_hspace = 0.25
inner_wspace = 0
inner_hspace = 0.22
bins = 100
xlims = (-180, 180)
ylims = (65,175)
xlines = range(-150,151,50)
ylines = range(80,161,20)

d_axes = p_agent.histogram2d_chi_delta_four_systems(figsize, outer_wspace, outer_hspace, inner_wspace, inner_hspace, bins, xlims, ylims, xlines, ylines)

plt.savefig("/home/yizaochen/Desktop/drawzone_temp/chi-delta-four-systems.png", dpi=150, transparent=False)
plt.show()
```

```python
host = 'tat_21mer'
strand_id = 'STRAND2'
bp_list_junction = [10, 11, 12]
bp_list_non_junction = list(range(4, 10)) + list(range(13, 19))

chi_junction, delta_junction = p_agent.get_data_chi_delta_by_bplist(strand_id, host, bp_list_junction)
chi_nonjunction, delta_nonjunction = p_agent.get_data_chi_delta_by_bplist(strand_id, host, bp_list_non_junction)
```

```python
bins = 100
xlims = (-180, 180)
ylims = (65,175)
xlines = range(-150,151,50)
ylines = range(80,161,20)
fig, axes = plt.subplots(nrows=2, figsize=(8,8), facecolor="white")

ax = axes[0]
ax.hist2d(chi_junction, delta_junction, bins=bins, density=True, cmap='Reds')
ax.set_xlim(xlims)
ax.set_ylim(ylims)
for xvalue in xlines:
    ax.axvline(xvalue, color="grey", alpha=0.1)
for yvalue in ylines:
    ax.axhline(yvalue, color="grey", alpha=0.1)
ax.set_xlabel("$\chi$")
ax.set_ylabel("$\delta$")
ax.set_title('A-junction Strand2 Junction Part')

ax = axes[1]
ax.hist2d(chi_nonjunction, delta_nonjunction, bins=bins, density=True, cmap='Reds')
ax.set_xlim(xlims)
ax.set_ylim(ylims)
for xvalue in xlines:
    ax.axvline(xvalue, color="grey", alpha=0.1)
for yvalue in ylines:
    ax.axhline(yvalue, color="grey", alpha=0.1)
ax.set_xlabel("$\chi$")
ax.set_ylabel("$\delta$")
ax.set_title('A-junction Strand2 Non-Junction Part')

plt.tight_layout()
#plt.savefig("/home/yizaochen/Desktop/drawzone_temp/A_junction_chi_delta.png", dpi=150, transparent=False)
plt.show()
```
